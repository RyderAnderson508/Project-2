```python
#Import necessary packages
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Assumptions and Constants 
# CUBIC_YARD_FT3 = 27       # cubic feet in one cubic yard
# UNIT_WEIGHT_WATER = 62.4  # lb/ft^3 (unit weight of water)
# The division of the target percentages by 100 is to go from percent to decimal form.
```

```python
#Code Dictionary 
# A  = Weight of cement per cubic yard (lb)
# B  = Weight of fly ash per cubic yard (lb)
# C  = Weight of silica fume per cubic yard (lb)
# D  = Weight of other SCM per cubic yard (lb)
# E  = Target water–cement ratio (decimal)
# F  = Target percent air content (%)
# G  = Target percent fine aggregate (%)
# H  = Target percent coarse aggregate (%)
# I  = Target percent other aggregate (%)

# J  = Specific gravity of cement
# K  = Specific gravity of fly ash
# L  = Specific gravity of silica fume
# M  = Specific gravity of other SCM
# N  = Specific gravity of fine aggregate
# O  = Specific gravity of coarse aggregate
# P  = Specific gravity of other aggregate

# Q  = Weight of water (lb), computed as (A + B + C + D) * E
# R  = Volume of cement (ft^3)
# S  = Volume of fly ash (ft^3)
# T  = Volume of silica fume (ft^3)
# U  = Volume of other SCM (ft^3)
# V  = Volume of air (ft^3)
# W  = Volume of water (ft^3)
# X  = Total aggregate volume required (ft^3)
# Y  = Weight of fine aggregate (lb)
# Z  = Weight of coarse aggregate (lb)
# AA = Weight of other aggregate (lb)

```

```python
#User Input: Project Information
project_number = int(input("Enter project number: "))
class_of_concrete = input("Enter class of concrete (e.g., 47B, 47BR): ")

# User Input: Target Amounts
A = int(input("Enter the weight of cement per cubic yard:"))
B = int(input("Enter the weight of fly ash in cubic yard"))
C = int(input("Enter the weight of silica fume per cubic yard:"))
D = int(input("Enter the weight of SCM per cubic yard:"))
E = float(input("Enter the target water/cement ratio:"))
F = int(input("Enter the target % of air content:"))
G = int(input("Enter the target % of fine aggregate:"))
H = int(input("Enter the target % of coarse aggregate:"))
I = int(input("Enter the target % of other aggregate:"))

# User Input: Specific Gravity Values
J = float(input("Enter the specific gravity of cement:"))
K = float(input("Enter the specific gravity of fly ash:"))
L = float(input("Enter the specific gravity of silica fume:"))
M = float(input("Enter the specific gravity of other SCM:"))
N = float(input("Enter the specific gravity of fine aggregate:"))
O = float(input("Enter the specific gravity of coarse aggregate:"))
P = float(input("Enter the specific gravity of other aggregate:"))

# Create a function to automate the mix design
def mix_design(A, B, C, D, E, F, G, H, I, J, K, L, M, N, O, P):
    Q = (A + B + C + D) * E # weight of water
    R = A / (J*62.4) #volume of cement (ft^3)
    S = B / (K*62.4) #volume of flyash (ft^3)
    T = C / (L*62.4) # volume of silica fume (ft^3)
    U = D / (M*62.4) # volume of other SCM (ft^3)
    V = (F/100) * 27 # volume of air (ft^3)
    W = Q / 62.4 # volume of water (ft^3)
    X = 27 - R - S - T - U - V - W # volume of total aggregate
    Y = (62.4 * (G/100)*N*X) #weight of fine aggregate
    Z = (62.4 * (H/100)*O*X) # weight of coarse aggregate
    AA = (62.4 * (I/100)*P*X) # weight of other aggregate
    return(A, B, C, D, Y, Z, AA, Q, E, F)

# Utilize Function for Mix Design
cement, fly_ash, silica_fume, other_scm, fine_aggregate, coarse_aggregate, other_aggregate, water, wc_ratio, air = mix_design(A, B, C, D, E, F, G, H, I, J, K, L, M, N, O, P)

# Print Target Mix Design
print("\n---------------------------------------------")
print(" NDOT Concrete Mix Design – Weight Summary")
print("         (1 Cubic Yard of Concrete)")
print("---------------------------------------------")
print(f"Project Number:       {project_number}")
print(f"Class of Concrete:    {class_of_concrete}")
print("---------------------------------------------")
print(f"Cement (A):            {cement:8.1f} lb")
print(f"Fly Ash (B):           {fly_ash:8.1f} lb")
print(f"Silica Fume (C):       {silica_fume:8.1f} lb")
print(f"Other SCM (D):         {other_scm:8.1f} lb")
print("---------------------------------------------")
print(f"Fine Aggregate (Y):    {fine_aggregate:8.1f} lb")
print(f"Coarse Aggregate (Z):  {coarse_aggregate:8.1f} lb")
print(f"Other Aggregate (AA):  {other_aggregate:8.1f} lb")
print("---------------------------------------------")
print(f"Water Weight(Q):       {water:8.1f} lb")
print(f"Water/Cement Ratio:    {wc_ratio:8.1f}")
print(f"Air Content:           {air:8.1f} %")
print("---------------------------------------------")
print("End of Mix Design Summary")
print("Contractor Target Mix Design Weights for 1 Cubic Yard:")
```
